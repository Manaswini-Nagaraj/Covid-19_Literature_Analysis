```python

import numpy as np
import pandas as pd 
import re
import nltk
from nltk.corpus import stopwords, comparative_sentences
from collections import Counter
import os
from nltk import bigrams
```

```python
texts = lst
```

```python
stopwords_eng = stopwords.words("english")
```

```python
text = "".join([x[:1000] for x in texts])
relev_words = [x for x in text.split(" ") if x.lower() not in stopwords_eng and len(x) >= 3]
relev_text = " ".join([x.lower() for x in relev_words])
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline



wordcloud = WordCloud(width = 800, height = 600, 
                background_color ='black', 
                stopwords = stopwords_eng, 
                min_font_size = 10, max_words = 100).generate(relev_text) 
  
                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud, interpolation= "bilinear") 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
```
